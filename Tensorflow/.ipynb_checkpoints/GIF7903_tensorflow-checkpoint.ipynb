```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Chargement et prétraitement des données
train_datagen = ImageDataGenerator(rescale=1./255)
validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    'dataset\\train',  # dataset entraînement
    target_size=(48, 48),  
    batch_size=20,
    class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
    'dataset\\validation',  # dataset validation
    target_size=(48, 48),
    batch_size=20,
    class_mode='binary')


model_path = 'trained_model.tf'

# Vérifier si le modèle existe déjà
if os.path.exists(model_path):
    print("Le modèle existe.")
    choice = input("Appuyez sur '1' pour ré-entraîner le modèle, sinon appuyez sur n'importe quelle autre touche: ")
    if choice == '1':
        print("Reprise de l'entraînement...")
        model_tf = load_model(model_path)
        model_tf.fit(
            train_generator,
            steps_per_epoch=7, 
            epochs=15, 
            validation_data=validation_generator,
            validation_steps=1)
        model_tf.save(model_path)
        print("Le modèle a été ré-entraîné et sauvegardé.")
    else:
        model_tf = load_model(model_path)
        print("Le modèle existant n'a pas été modifié.")
else:
    print("Création du modèle...")
    # Definition du modèle
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(48, 48, 3)),
        MaxPooling2D(2, 2),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),
        Flatten(),
        Dense(512, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    # Compilation du modèle
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    # Entraînement du modèle
    modele_tf = model.fit(
        train_generator,
        steps_per_epoch=7,  # 7.2x20= 144 (taille de la dataset entraînement)
        epochs=15,
        validation_data=validation_generator,
        validation_steps=1)  
    
    # Sauvegarde du modèle
    model.save(model_path)
    print("Le modèle a été ré-entraîné et sauvegardé.")
```

```python
# Fonction pour charger et prétraiter une image
def load_and_preprocess_image(image_path):
    img = image.load_img(image_path, target_size=(48, 48))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0
    return img_array

# Chargement de l'image 
image_test = 'arbre_trou.jpg'
test_image = load_and_preprocess_image(image_test)

# Affichage de l'image 
plt.imshow(plt.imread(image_test)) 
plt.title(os.path.basename(image_test)) 
plt.show()

# Prédiction avec le modèle existant
prediction = model.predict(test_image)

# Interprétation de la prédiction
if prediction[0][0] > 0.5:
    result = "Il y'a un (ou plusieurs) de trous sur le tronc"
else:
    result = "Il n y'a pas de trous sur le tronc"
print(result)
```
